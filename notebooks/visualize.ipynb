```python
### Comparizons::

comparison = pd.DataFrame({
    "dataset": ["nfcorpus", "fiqa", "quora", "nq"],

    "paper_bm25": [
        0.322,
        0.236,
        0.789,
        0.305
    ],

    "paper_splade": [
        0.347,
        0.347,
        0.834,
        0.538
    ],

    "my_bm25": [
        0.32178,
        0.23607,
        0.78859,
        0.30548
    ],

    "my_splade": [
        0.34731,
        0.34728,
        0.83435,
        0.53730
    ]
})
```

```python
### Reproduction error:

comparison["bm25_diff"] = (
    comparison["my_bm25"]
    - comparison["paper_bm25"]
)

comparison["splade_diff"] = (
    comparison["my_splade"]
    - comparison["paper_splade"]
)

comparison["bm25_abs_error_%"] = (
    comparison["bm25_diff"].abs()
    / comparison["paper_bm25"]
    * 100
)

comparison["splade_abs_error_%"] = (
    comparison["splade_diff"].abs()
    / comparison["paper_splade"]
    * 100
)

print(comparison.round(4))
```
